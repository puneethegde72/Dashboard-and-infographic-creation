```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Used Cars Dataset from Kaggle
df = pd.read_csv('matches.csv')
```

```python
print("Number of matches played:", len(df))
print("Number of teams:", len(df['team1'].unique()))
print("Number of seasons:", len(df['Season'].unique()))
print("Most man of the match awards:", df['player_of_match'].value_counts().idxmax())
```

```python
# Visualization 1: Number of matches played by season
plt.figure(figsize=(10,5))
sns.countplot(x='Season', data=df, palette='viridis')
plt.title("Number of matches played by season")
plt.xlabel("Season")
plt.ylabel("Number of matches")
plt.savefig('your-student-id_1.png', dpi=300)
```

```python
# Visualization 2: Result of matches by toss decision
plt.figure(figsize=(10,5))
sns.countplot(x='toss_decision', hue='result', data=df, palette='coolwarm')
plt.title("Result of matches by toss decision")
plt.xlabel("Toss decision")
plt.ylabel("Number of matches")
plt.legend(title="Result")
plt.savefig('your-student-id_2.png', dpi=300)
```

```python
# Visualization 3: Number of matches played by city
plt.figure(figsize=(10,10))
sns.countplot(y='city', data=df, palette='magma', order=df['city'].value_counts().index)
plt.title("Number of matches played by city")
plt.xlabel("Number of matches")
plt.ylabel("City")
plt.savefig('your-student-id_3.png', dpi=300)
```

```python
# Visualization 5: Distribution of total runs in each inning
innings = df.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.kdeplot(x='total_runs', hue='inning', data=innings, palette='crest', common_norm=False, fill=True, alpha=0.5, linewidth=0)
plt.title("Distribution of total runs in each inning")
plt.xlabel("Total runs")
plt.ylabel("Density")
plt.legend(title="Inning")
plt.savefig('your-student-id_5.png', dpi=300)
```
